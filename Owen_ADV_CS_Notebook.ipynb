```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np



# Start coding here!
#creating dataframe from dataset nobel
#useful for plotting
import matplotlib.pyplot as plt
nobel = pd.read_csv('data/nobel.csv')

#creating variables to store data on most common awarded gender and birth country, accessing dataframe column, finding the most common variable with mode which returns a series, with the largest number in the first spot which can be accessed using iloc[0]

top_gender = nobel['sex'].mode().iloc[0]
top_country = nobel['birth_country'].mode().iloc[0]

#creates a column with all US winners

nobel['USA_winner'] = nobel['birth_country'] == 'United States of America'

#creates another column with every decade but as an integer

nobel['decade'] = np.floor(nobel['year'].astype(int)/10)*10

#assigns the highest ratio decade to max_decade usa by first grouping the nobel prize winners by decade, then makes sure it is a dataframe. Afterwards subsets the US born winners, and gets the mean for each decade. next sorts of the mean so the highest value is at the top, and  accesses this value with iloc.
max_decade_usa = nobel.groupby('decade',as_index = False)['USA_winner'].mean().sort_values('USA_winner', ascending = False).astype(int).iloc[0,0]
#plots the data for decades and proportion of winners
sns.relplot(kind='line', x='decade',y='USA_winner',data=nobel,errorbar = None)
#makes data easier to read
plt.xlabel('Time(Decades)')
plt.ylabel('Prop. of US Born Winners')
plt.title('Proportion of US Born Nobel Prize Winners Over Time')
plt.show()
#creates a column with all Female winners
nobel['female_winner'] = nobel['sex'] == 'Female'

#same code as the one for finding the max_decade_usa except without finding the exact one, just finding proportions after grouping by decade and category
female_prop = nobel.groupby(['decade','category'],as_index = False)['female_winner'].mean()
#had to get ai to help for this one, basically was trying to do it in one line and was using the value of the proportion to locate the both decade and category which did not work. This approach takes the proportion, then finds the row where the proportion is equal to the maximum proportion
max_female_prop = female_prop['female_winner'].max()
max_female_rows = female_prop[female_prop['female_winner'] == max_female_prop]
max_female_dict = { max_female_rows['decade'].values[0]:max_female_rows['category'].values[0]
}
#creating the plot, removing the error from being shown, and seperating into category
sns.relplot(kind='line',data=female_prop,errorbar=None,x='decade',y='female_winner',hue = 'category' )
#cleans up labels making it easier to read
plt.xlabel("Time(Decades)")
plt.ylabel("Prop.of Female Winners")
plt.title('Proportion of Female Nobel Prize Winners Over Time')
plt.show()

#finding the first year that a woman won
first_woman_year = nobel[nobel['female_winner']]['year'].min()
first_woman = nobel[(nobel['year'] == first_woman_year) & (nobel['sex'] == 'Female') ]
#storing the required parts as strings
first_woman_category = first_woman.astype(str)['category'].iloc[0]
first_woman_name = first_woman.astype(str)['full_name'].iloc[0]

#counts repeated names
count = nobel['full_name'].value_counts()
#takes the data series of counts and saves the names of counts greater than 1 to the repeat list.
repeat_list = count[count>1].index.tolist()

```
