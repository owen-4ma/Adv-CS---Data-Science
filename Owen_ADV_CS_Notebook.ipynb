- pandas package
  - built on the numpy function but is more advanced
  - stores data in something called a dataframe
  - import the package using import pandas as pd
  - can create dataframes from dictionaries but also can import the data externally and create it on its own ex: a CSV (comma seperated variables) file
  - when creating manually use name of data frame = pd.DataFrame(dictionary name)

This data is set as a pandas series, which is like a 1d array (data frame is like multiple series in one object) if you want the data type to be a dataframe, use double square brackets ex: brics[['country]] you can add a comma and two column labels to create a sub dataframe with only those two columns. you can also slice using the same principles
  - advanced data access methods: square brackets are limited, so to get something that is efficient like 2d arrays, you use functions like loc and iloc
    - loc: can access rows using dataframe.loc['rowlabel'] again, to get a dataframe just double the square brackets. by adding commas to the double square brackets you can access multiple rows at the same time. you can also access only certain rows and certain columns by using two square brackets for example, dataframe.loc[["rowlabel1", "rowlabel2'], ["columnlabel1","columnlabel2"] if you want all rows but only use specific columns,slice and use dataframe.loc[[:, ["columnlabel1","columlabel2"]  

    - iloc: based on positions, pretty intuitive basically like 2d arrays. you can still slice. they are pretty similar to loc just without the label names and using index instead
    - summary of differences: square brackets allow for column access, and row access is soley through slicing. loc allows for label based access to rows first, and then columns as well allthough rows the entirety of rows has to be sliced first (using :,) iloc allows for 2d array like access using positions

**Data Manipulation with pandas**
- pandas is built upon numpy and matplotlib
- rectangular/tabular data is most common
  - represented as a dataframe which is used almost everywhere
- Methods/functions:
  - .head() returns the first few rows (the “head” of the DataFrame).
  - .info() shows information on each of the columns, such as the data type and number of missing values.
  - .shape returns the number of rows and columns of the DataFrame.
  - .describe() calculates a few summary statistics for each column.
  - .values: A two-dimensional NumPy array of values.
  - .columns: An index of columns: the column names.
  - .index: An index for the rows: either row numbers or row names.
  - .sort_values() will sort through values in a given column defaults to ascending however you can add the arguement that sets ascending to false which will then sort them in descending order. can further make it a list and sort through multiple values. One column is just one square bracket, while two columns is multiple.
- python has a philosophy that there is one obvious solution citing 'the zen of python', however pandas deliberately doesnt follow this philosophy and will frequently have multiple correct solutions
- Subsetting
  - ex for one named icecreams you will use icecream['name']
  - one square bracket for a single column
  - two square bracket for multiple (instead would be like icecream[['name','name2']]
  - can also use a variable like (var = ['name', 'name2'] icecream[var])
  - can also filter through rows with logical conditions like in a hypothetical involving dogs dogs[dogs[height_cm] > 50] would return all dogs that have a hieght above fifty centimeters.
  - can also filter witha. double equals sign and text data
  - if involvings dates remember to use the international date standard (year, month, date)
  - can use multiple conditions but unless using variables will require parentheses
  - can also use isin() method which takes multiple values to filter for
  - editing a dataframe can be referred to as "mutating", "transforming", and "Feature Engineering"
  - all you have to do to add a column is to put a new label as an arguement in square brackets ex: icecream['flavor'] = _calculation here_
