- Objects have methods, basically functions assigned to specific objects, called by adding a .methodName to the end of the object, for example **str.Upper()**.

- packages, directory of different python scripts (called modules), which specify functions, methods, and type.
  - For Data Science, there are packages like: NumPy, Matplotlib, sikit-learn.
  - To use packages, you have to install them using terminal, then import by using the import statement followed by the package. you can also refer to the package as something else using as. EX: Import NumPy as np.
  - If you only want to import one aspect like a function, use the from statement + packageName in front of the import statement+functionName, EX: from NumPy import array
- NumPy
  - an alternative to the default python lsit called the NumPy array
  - added features like calculation over whole arrays
  - easy and fast
  - it assumes all values are of the same type if there are multiple types, it will default to strings
  - has unique methods that might be difficult for example, normally you can combine to lists by using + but in NumPy it will return the element-wise sum of the arrays. Another example is booleans true and false will be converted to 1 and 0 respectively
  - the unique ways of subsetting like the greater than sign
  - 2D arrays also possible, just like 1D 2D is like an improved version, a list of lists.
  - enables better subsetting, u can also use single brackets + comma for more efficiency
  - Creating one is just name = np.array(values/already existing array)

  - **Corrcoef** is a function that can check fi two variables are correlated
